```python
import pandas as pd
import hvplot.pandas
import matplotlib.pyplot as plt
import panel as pn
import seaborn as sns
datasets =  pn.widgets.Select(name='Datasets',options=['cancer-incidence.csv','number-of-people-with-cancer.csv','cancer-deaths-by-age.csv','cancer-deaths-by-type.csv','five-year-survival-rates-by-cancer-type.csv'])
kind = pn.widgets.Select(name='kind',options=['violin','line','step','barh','bar'])
country = pn.widgets.Select(name='country',options=['China','United States','India','Iraq','Mongolia','South Africa'])
@pn.depends(datasets,kind,country)
def make_plot(datasets,kind,country):
    df_plot = pd.read_csv(datasets)
    if datasets=='cancer-incidence.csv':
        df_plot.columns = ['Entity','Code','Year','Incidence']
        Incidence_data_choose_country = df_plot[df_plot['Entity'].isin(['China','United States','India','Iraq','Mongolia','South Africa'])]
        if kind=='violin':
            plot = Incidence_data_choose_country.hvplot.violin(y='Incidence',group_label='Country',value_label='Cancer incidence',invert=True,color='Entity',by='Entity',width=1000,height=500)
            return plot
        if kind=='line':
            plot = Incidence_data_choose_country.hvplot.line(x='Year',y='Incidence',by='Entity',height=500,width=1000)
        if kind=='step':
            plot = Incidence_data_choose_country.hvplot.step(x='Year',y='Incidence',by='Entity',height=500,width=1000)
        return plot
    if datasets=='number-of-people-with-cancer.csv':
        df_plot.columns = ['Entity','Code','Year','Prevalence']
        number_data_choose_contry = df_plot[df_plot['Entity'].isin(['China','United States','India','Iraq','Mongolia','South Africa'])]
        if kind=='barh':
            plot = number_data_choose_contry.hvplot.barh('Year','Prevalence',by='Entity',stacked=True,height=500,width=1000)
        return plot
    if datasets=='cancer-deaths-by-age.csv':
        df_plot.columns = ['Entity','Code','Year','>70','50-69','15-49','5-14','<5']
        age_data_choose_contry = df_plot[df_plot['Entity'].isin(['China','United States','India','Iraq','Mongolia','South Africa'])]
        age_data_choose_contry_2019 = age_data_choose_contry[age_data_choose_contry['Year']==2019]
        age = age_data_choose_contry_2019.columns[3:].tolist()
        plot = age_data_choose_contry_2019.hvplot.bar(x='Entity',y=age,ylabel='Deaths',xlabel='Entity',width=1500,height=1000)
        return plot
    if datasets=='cancer-deaths-by-type.csv':
        df_plot.columns = ['Entity','Code','Year','lung','Cervical','Breast','Brain','Other','Bladder','lymphoma','myeloma','skin','Esophageal','Pancreatic','Larynx','Prostate','Uterine','Hodgkin','Thyroid','Stomach','Colon','oral cavity','pharynx','Non-melanoma','Leukemia','Liver','Ovarian','Gallbladder','Nasopharynx','Mesothelioma','Kidney','Testicular']
        type_data_choose_contry = df_plot[df_plot['Entity'].isin(['China','United States','India','Iraq','Mongolia','South Africa'])]
        type_data_choose_contry_2019 = type_data_choose_contry[type_data_choose_contry['Year']==2019]
        if country=='United States':
            type_data_United_States_2019 = type_data_choose_contry_2019[type_data_choose_contry_2019['Entity']=='United States']
            type = type_data_choose_contry_2019.columns[3:].tolist()
            plot = type_data_United_States_2019.hvplot.bar(x='Entity',y=type,ylabel='Deaths',xlabel='Cancer type',width=2000,height=1000,title='United States')
            return plot
        if country=='China':
            type_data_China_2019 = type_data_choose_contry_2019[type_data_choose_contry_2019['Entity']=='China']
            type = type_data_choose_contry_2019.columns[3:].tolist()
            plot = type_data_China_2019.hvplot.bar(x='Entity',y=type,ylabel='Deaths',xlabel='Cancer type',width=2000,height=1000,title='China')
            return plot
        if country=='India':
            type_data_India_2019 = type_data_choose_contry_2019[type_data_choose_contry_2019['Entity']=='India']
            type = type_data_choose_contry_2019.columns[3:].tolist()
            plot = type_data_India_2019.hvplot.bar(x='Entity',y=type,ylabel='Deaths',xlabel='Cancer type',width=2000,height=1000,title='India')
            return plot
        if country=='Iraq':
            type_data_Iraq_2019 = type_data_choose_contry_2019[type_data_choose_contry_2019['Entity']=='Iraq']
            type = type_data_choose_contry_2019.columns[3:].tolist()
            plot = type_data_Iraq_2019.hvplot.bar(x='Entity',y=type,ylabel='Deaths',xlabel='Cancer type',width=2000,height=1000,title='Iraq')
            return plot
        if country=='Mongolia':
            type_data_Mongolia_2019 = type_data_choose_contry_2019[type_data_choose_contry_2019['Entity']=='Mongolia']
            type = type_data_choose_contry_2019.columns[3:].tolist()
            plot = type_data_Mongolia_2019.hvplot.bar(x='Entity',y=type,ylabel='Deaths',xlabel='Cancer type',width=2000,height=1000,title='Mongolia')
            return plot
        if country=='South Africa':
            type_data_South_Africa_2019 = type_data_choose_contry_2019[type_data_choose_contry_2019['Entity']=='South Africa']
            type = type_data_choose_contry_2019.columns[3:].tolist()
            plot = type_data_South_Africa_2019.hvplot.bar(x='Entity',y=type,ylabel='Deaths',xlabel='Cancer type',width=2000,height=1000,title='South Africa')
            return plot
    if datasets=='five-year-survival-rates-by-cancer-type.csv':
        survival_rate_choose_country = df_plot[df_plot['Entity'].isin(['China','United States','India','Iraq','Mongolia','South Africa'])]
        survival_rate_choose_country_2019 = survival_rate_choose_country[survival_rate_choose_country['Year']==2009]
        cancer_type = survival_rate_choose_country_2019.columns[3:].tolist()
        fig,ax = plt.subplots(figsize=(8,8))
        ax.plot(cancer_type,survival_rate_choose_country_2019.iloc[0,3:].tolist(),label='China')    # 中国
        ax.plot(cancer_type,survival_rate_choose_country_2019.iloc[1,3:].tolist(),label='India')    # 印度
        ax.plot(cancer_type,survival_rate_choose_country_2019.iloc[2,3:].tolist(),label='Mongolia') # 蒙古
        ax.plot(cancer_type,survival_rate_choose_country_2019.iloc[3,3:].tolist(),label='South Africa') # 南非
        ax.plot(cancer_type,survival_rate_choose_country_2019.iloc[4,3:].tolist(),label='United States')# 美国
        ax.legend()
        return fig
template = pn.template.FastListTemplate(
    site='MED5018',title='第四小组画图展示',
    sidebar=["## Settings",datasets,kind,country],
    main=['Main plot',make_plot]).servable()
```
